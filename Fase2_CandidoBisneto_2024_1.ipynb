---

**Descrição:**
\
Na manipulação de strings e algoritmos de comparação, uma tarefa comum é verificar se duas palavras ou frases contêm os mesmos caracteres na mesma quantidade, independentemente da ordem. Esta habilidade é útil em vários domínios da computação, como criptografia, recuperação de informação e processamento de linguagem natural.

\\

**Tarefa:**
\
Desenvolva uma função chamada ***verificar_permutacao(palavra1, palavra2)*** que determina se as duas strings fornecidas são permutações uma da outra. A função deve retornar True se as strings contiverem os mesmos caracteres com as mesmas contagens de caracteres, e False caso contrário.

* A função deve tratar caracteres maiúsculos e minúsculos como equivalentes e ignorar espaços e pontuações.
* Considerar que strings que diferem em tamanho não podem ser permutações uma da outra.

\\
Exemplos:

* "cat" e " tAc" retorna-se True, pois desconsiderando pontos, espaços e capitalização, as palavras são permutações uma da outra.
* "jack" e "black" retorna-se False
* "Bola" e "alob" retorna-se True

```python
def verificar_permuta(texto1, texto2) -> bool:

  # verifica se os textos tem tamanhos iguais
  if len(texto1) != len(texto2):
    return False

  # validação de semelhança entre os textos pelo seu conteudo
  # texto 2 é invertido e após isso são comparados, caso sejam
  # iguias desconciderando letras maiúsculos e minúsculos retorna
  # True
  if len(texto1) == len(texto2):
    texto2_invertido = texto2[::-1]
    if texto1.casefold() == texto2_invertido.casefold():
      return True
    else: return False
```

Testando

```python
texto1 = "cat"
texto2 = "Tac"

verificar_permuta(texto1, texto2)
```

**Descrição:**
\
Em muitos sistemas, é comum a necessidade de buscar elementos em uma coleção de dados. Entender a eficiência de diferentes algoritmos de busca é essencial para otimizar o desempenho do sistema.

\\

**Tarefa:**
\
Implemente duas funções de busca que encontrem um elemento em uma lista não ordenada: \\
* busca_linear(lista, elemento) que utiliza a busca linear para encontrar um elemento na lista. \\
* busca_ordenada(lista, elemento) que primeiramente ordena a lista e depois realiza uma busca binária para encontrar o elemento. \\

\\

**Requisitos:**


* As funções devem retornar o índice do elemento na lista se ele estiver presente, ou -1 se ele não estiver.
* Para o caso da busca ordenada, retorne o novo indice após a ordenação.

\\
Exemplos:

* lista=[0, 3, 2, 5], elemento=3, retorna-se 1
* lista=[1, 2, 8, 90], elemento=3, retorna-se -1
* lista=[55, 30, 23, 12, 5, 67, 199], elemento=67, retorna-se 5

\\
Referências:

[1] https://pt.wikipedia.org/wiki/Pesquisa_bin%C3%A1ria

```python
def busca_linear(lista, elemento) -> int:
  for elementos in lista:
    # Verifica se o elemento atual do loop é igual ao elemento
    # procurado. Se sim, é retornado seu index, caso contário, -1
    if elemento == elementos:
      return lista.index(elementos)
  return -1
```

```python
def busca_ordenada(lista, elemento) -> int:
  start = 0
  end = len(lista) - 1
  lista.sort()

  #  Inicia um loop que continuará enquanto o índice inicial for menor ou
  # igual ao índice final. Isso garante que a busca continue até que o
  # intervalo seja reduzido a um único elemento.
  while start <= end:
    meio = (start + end) // 2
    if lista[meio] == elemento:
       return lista.index(elemento)
    elif lista[meio] < elemento:
        # e o elemento do meio for menor que o elemento procurado,
        # atualiza o índice inicial para ser o próximo elemento após o meio.
        start = meio + 1
    else:
      # Se o elemento do meio for maior que o elemento procurado,
      # atualiza o índice final para ser o elemento anterior ao meio.
      end = meio - 1

  return -1
```

```python
def criptografar_cifra_vigenere(mensagem, chave) -> str:
    chave = chave.upper()
    mensagem_criptografada = ''

    for i, letra in enumerate(mensagem):
        letra = letra.upper()

        # Verificar se a letra é uma letra do alfabeto
        if letra.isalpha():
            # Calcula o deslocamento com base na letra atual da palavra-chave.
            deslocamento_chave = ord(chave[i % len(chave)]) - ord('A')
            # Calcula a letra criptografada com base na letra atual da mensagem
            # e no deslocamento da palavra-chave. A letra é convertida para seu
            # valor ASCII, o deslocamento é aplicado, e então o resultado é convertido
            # de volta para uma letra usando a função chr()
            letra_criptografada = chr((ord(letra) - ord('A') + deslocamento_chave) % 26 + ord('A'))
            mensagem_criptografada += letra_criptografada
        else:
            mensagem_criptografada += letra

    return mensagem_criptografada
```

```python
def descriptografar_cifra_vigenere(mensagem, chave) -> str:
    chave = chave.upper()
    mensagem_descriptografada = ''

    for i, letra in enumerate(mensagem):
        letra = letra.upper()

        if letra.isalpha():
          #  Calcula o deslocamento na cifra de Vigenère para esta posição na
          # mensagem, usando a fórmula ord(chave[i % len(chave)]) - ord('A'),
          # onde ord() retorna o valor ASCII da letra e i % len(chave) garante
          #  que a chave seja repetida para mensagens mais longas que a chave.
            deslocamento_chave = ord(chave[i % len(chave)]) - ord('A')
            # Descriptografa a letra atual usando a fórmula da Cifra de Vigenère, que
            # é ((valor_letra - deslocamento_chave) % 26 + valor_A), onde valor_letra é
            # o valor ASCII da letra, deslocamento_chave é o deslocamento calculado da chave
            letra_descriptografada = chr((ord(letra) - ord('A') - deslocamento_chave) % 26 + ord('A'))
            mensagem_descriptografada += letra_descriptografada
        else:
            mensagem_descriptografada += letra

    return mensagem_descriptografada
```

# Questão 4: Validação de CPF
---

**Descrição:**


Você foi encarregado de desenvolver uma funcionalidade crítica para o sistema de cadastro de clientes de uma aplicação bancária. Esta funcionalidade envolve a validação da autenticidade de números de Cadastro de Pessoa Física (CPF), um documento essencial para identificação de indivíduos no Brasil, bem como a identificação do estado de origem desse CPF.

\\

**Tarefa:**



*   Implemente uma função ***validar_cpf(cpf)*** que receba uma string contendo um número de CPF e retorne True se o CPF for autêntico e False caso contrário.
*   Implemente uma função ***identificar_estado(cpf***) que retorne o estado de origem do CPF, caso o número seja autêntico. Se o CPF não for autêntico ou não tiver um estado identificável, a função deve retornar ***None.***


**Requisitos:**


*   O algoritmo de validação deve considerar os dígitos verificadores do CPF conforme as regras matemáticas de sua composição.
*   A função de identificação do estado deve considerar o nono dígito do CPF, conforme a tabela de alocação de números por estado.

*   As funções devem ser capazes de processar CPFs com ou sem pontuação (pontos e traço).




\\
Referências:

[1] http://clubes.obmep.org.br/blog/a-matematica-nos-documentos-a-matematica-dos-cpfs/

```python
def validar_cpf(cpf) -> bool:

  cpf_n = cpf.replace('.', '').replace('-', '')

  # Verifica se o número de CPF após a remoção dos pontos e
  # traço tem exatamente 11 dígitos
  if len(cpf_n) != 11:
    return False

  soma = 0
  for i in range(9):
    # Calcula a soma ponderada dos dígitos, multiplicando cada dígito
    # pelo seu peso correspondente e somando-os à variável
	  soma += int(cpf_n[i]) * (10 - i)
  if soma % 11 < 2:
	  primeiro_digito = 0
    # Calcula o primeiro dígito verificador do CPF com base
    # no resultado da verificação anterior.
  else: primeiro_digito = 11 - (soma % 11)

  # Verifica se o primeiro dígito verificador do CPF fornecido é igual ao
  # primeiro dígito verificador calculado. Se não for, o CPF é inválido e
  # a função retorna False
  if int(cpf_n[9]) != primeiro_digito:
    return False

  # O processo é repetido para calcular e verificar o segundo dígito verificador,
  # usando os 10 primeiros dígitos do CPF. Se os dois dígitos verificadores
  # calculados coincidirem com os dígitos verificadores fornecidos no CPF, a
  # função retorna True, indicando que o CPF é válido
  soma = 0
  for i in range(10):
    soma += int(cpf_n[i]) * (11 - i)
  if soma % 11 < 2:
    segundo_digito = 0
  else: segundo_digito = 11 - (soma % 11)

  if int(cpf_n[10]) != segundo_digito:
    return False

  return True
```

```python
def identificar_estado(cpf) -> str:

  cpf_n = cpf.replace('.', '').replace('-', '')

  if len(cpf_n) != 11:
    return "CPF Inválido"

  # Define um dicionário chamado dados que mapeia cada dígito do CPF
  # para uma lista de estados brasileiros associados a esse dígito.
  dados = {
    '1': ['DF', 'GO', 'MS', 'MT', 'TO'],
    '2': ['AC', 'AM', 'AP', 'PA', 'RO', 'RR'],
    '3': ['CE', 'MA', 'PI'],
    '4': ['AL', 'PB', 'PE', 'RN'],
    '5': ['BA', 'SE'],
    '6': ['MG'],
    '7': ['ES', 'RJ'],
    '8': ['SP'],
    '9': ['PR', 'SC'],
    '0': ['RS']
  }

  # Realiza a chamada para a obtenção do/dos estados relacionados a numeração
  # do 8º dígito do CPF
  estados = dados.get(cpf_n[8])
  return estados
```

### 7.2 - Detecção de Bordas
---
**Descrição:**
\
Dentro da visão computacional existe o conceito de bordas, que nada mais é do que determinar características de contornos de objetos em uma foto.

\\
**Tarefa:**
\
Tendo isso em mente, realize o processo de delimitação de bordas na imagem abaixo utilizando a função Canny Edge Detector do OpenCV.

```python
'''
O código abaixo converte a imagem carregada pela biblioteca Pillow para o
formato do OpenCV (numpy array).
'''

open_cv_image = convert_from_image_to_cv2(pil_img)
```

```python
# Aplicando filtro Gaussiano para melhorar a qualidade da imagem,
# a precisão da detecção de bordas e a localização das bordas.
blurred_img = cv2.GaussianBlur(open_cv_image, (5, 5), 0)
low_threshold = 50
high_threshold = 100

# Usando a função Canny para detectar as bordas
edges = cv2.Canny(blurred_img, low_threshold, high_threshold)
```

```python
'''
Depois de aplicar o filtro, converta de volta para o formato do Pillow
e printe a imagem.
'''

# Convertendo a imagem de bordas para o formato do Pillow
edges_image = convert_from_cv2_to_image(edges)

plt.subplot(121), plt.imshow(pil_img), plt.title('Imagem Original')
plt.subplot(122), plt.imshow(edges_image, cmap='gray'), plt.title('Imagem com Bordas')
plt.show()
```

### 8.1 Inicialização
* Defina os estados e ações: Considere 5 estados (0 a 4) e duas ações possíveis para cada estado (mover para a esquerda, mover para a direita).

* Inicialize a tabela Q com zeros: Crie uma matriz que represente a tabela Q, com linhas como estados e colunas como ações. Inicie todos os valores com zero.


**Parâmetros do Q-learning**


Taxa de aprendizado alpha(α): Determina o quanto a estimativa Q é atualizada em cada passo.

Fator de desconto gamma(γ): Avalia a importância das recompensas futuras.

Política
ϵ-greedy: Para equilibrar a exploração de novas ações com a exploração das já conhecidas.

```python
import numpy as np

# Definição dos parâmetros
alpha = 0.1
gamma = 0.9
epsilon = 0.2
states = [0, 1, 2, 3, 4]
actions = [0, 1] # 0: direita, 1: esquerda
q_table = np.zeros((len(states), len(actions)))
```

### 8.2 Escolhendo uma Ação
Implementação de uma função para escolher uma ação baseada na política
ϵ-greedy:

Com probabilidade ϵ, escolha uma ação aleatoriamente.
Com probabilidade 1−ϵ, escolha a ação com o maior valor Q para o estado atual.

**Recursos**

Towards Data Science - Epsilon Greedy Algorithm in Reinforcement Learning: Uma explicação detalhada da política
ϵ-greedy.

[1]https://towardsdatascience.com/tagged/epsilon-greedy-algorithm



```python
# Função para escolher a ação
def choose_action(state, q_table):
    # Exploration
    if np.random.uniform(0, 1) < epsilon:
        return np.random.choice(actions)
    # Explotation
    else:
        return np.argmax(q_table[state, :])
```

### 8.3 Atualizando a Tabela Q

Atualização da tabela Q usando a seguinte equação do Q-learning:
$$
Q(s, a) \leftarrow Q(s, a) + \alpha \left[ R(s, a) + \gamma \max_{a'} Q(s', a') - Q(s, a) \right]
$$
onde \( s \) é o estado atual, \( a \) é a ação tomada, \( R(s, a) \)  é a recompensa recebida, \( s' \) é o novo estado, e \( a' \) são as possíveis ações no novo estado.


**Recursos:**

Q-Learning Algorithm: A Comprehensive Guide from Scratch with Python Code: Um guia passo a passo para implementar Q-learning.

[1] https://www.analyticsvidhya.com/blog/2021/04/q-learning-algorithm-with-step-by-step-implementation-using-python/


```python
# Função para atualizar a Q-table
def update_q_table(state, action, reward, next_state, q_table):
    predict = q_table[state, action]
    target = reward + gamma * np.max(q_table[next_state, :])
    q_table[state, action] += alpha * (target - predict)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.arange(-5, 5.01, 0.01)
xs, ys = np.meshgrid(points, points)

ys
```

```python
z = np.sqrt(xs ** 2 + ys **2)
z
```

```python
plt.imshow(z, cmap=plt.cm.gray); 
plt.colorbar(); 
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")
```

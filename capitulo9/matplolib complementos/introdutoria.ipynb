```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])
```

```python
# Figure --> toda a imagem
fig = plt.figure() # figura vazia sem axes (eixos)
fig, ax = plt.subplots() # figura com um unico axes
fig, axs = plt.subplots(2, 2) # figura com uma grade 2x2 de eixo

# Figura com um axes na esquerda, e dois na direita
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

 - Axes --> é anexado a uma figura que contém uma região para 
plotagem de dados e geralmente inclui dois ou três Axis (cordenadas x, y, z)

 - Axis --> definem a escala e os limites e geram os traços nos eixos e etiquetas de identificação

 - Artist --> tudo o que é visível na figura (mesmo Figure, Axes e Axis), está ligado a um axes e não pode ser compartilhado por vários eixso ou movido de um para outro

```python
np.random.seed(19680801)
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter(x='a', y='b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

- Estilo de codificação

```python
# Ciração explicita de figuras e eixos chamando
# métodos neles usando OO

x = np.linspace(0, 2, 100)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('plotagem simples')
ax.legend()
```

```python
# Criação no estilo de pyplot

x = np.linspace(0, 2, 100)

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear') 
plt.plot(x, x**2, label='quadratic') 
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

- Fazendo funções auxiliar

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5,3))
my_plotter(ax1, data1, data2, {'marker': 'X'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
```

## Rotulagem de rotulo
- Etiquetas e texto dos eixos

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
n, bins, patches = ax.hist(x, bins=50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Comprimento em *cm*')
ax.set_ylabel('Probabilidade')
ax.set_title('Comprimento de um porco da terra\n')
ax.text(x=75, y=.025, s=r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

- Anotações

```python
# apontando uma seta para determinado lugar do gráfico
fig, ax = plt.subplots(figsize=(5, 3))
r = np.arange(0, 5, 0.01)
s = np.cos(2 * np.pi * r)
line = ax.plot(r, s, linewidth=1.5)

ax.annotate('local maximo', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
```

## Escalas axis e ticks
- Escalas

```python
#log
fig, axs = plt.subplots(1, 2, figsize=(5, 3), layout='constrained')
xdata = np.arange(len(data1))
data = 10**data1
axs[0].plot(xdata, data)

axs[1].set_yscale('log')
axs[1].plot(xdata, data)
```

Localizadores e formatadores de tick

```python
fig, axs = plt.subplots(2, 1, layout='constrained') 
axs[0].plot(xdata, data1)
axs[0].set_title('Ticks automático')

axs[1].plot(xdata, data1)
axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
axs[1].set_yticks([-1.5, 0, 1.5])
axs[1].set_title('Ticks manual')
```

Traçar datas e strings

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
dates = np.arange(np.datetime64('2021-11-15'), np.datetime64('2021-12-25'),
                  np.timedelta64(1, 'h'))
data = np.cumsum(np.random.randn(len(dates)))
ax.plot(dates, data)
cdf = mpl.dates.ConciseDateFormatter(ax.xaxis.get_major_locator())
ax.xaxis.set_major_formatter(cdf)
```

- Dados mapeados em cores

```python
X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

fig, axs = plt.subplots(2, 2, layout='constrained')
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

pc = axs[1, 0].imshow(Z**2 * 100, cmap='plasma',
                          norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

pc = axs[1, 1].scatter(data1, data2, c=data3, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')
```

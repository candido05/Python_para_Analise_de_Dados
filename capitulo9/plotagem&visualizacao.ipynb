Definindo o título, os rótulos dos eixoxs, os tiques e os rótulos dos tiques

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.standard_normal(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
                       rotation=30, fontsize=8)
ax.set_title('minha primiera plotagem')
ax.set_xlabel('Estagios')
```

```python
fig2 = plt.figure()
ay = fig2.add_subplot()
ay.hist(np.random.standard_normal(5000).cumsum(), bins = 50)
ticksY = ay.set_yticks([0, 100, 250, 300, 420, 500])
labelsY = ay.set_yticklabels(['0', '1K', '2K', '3K', '4K', '5K'],
                             rotation=30, fontsize=8)

props = {
    'title': 'plotagem com o eixo Y',
    'ylabel': 'Estagios',
}

ay.set(**props)
```

- Anotações e desenhos em uma subplotagem

```python
from datetime import datetime

fig = plt.plot()
ax = plt.subplot()
data = pd.read_csv('exemplos/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, color='#5a6ec4')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')
    
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Datas importantes na crise financeira de 2008-2009')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='#3e8749', alpha=0.5)
circ = plt.Circle((0.7, 0.2), 0.15, color='#ff9b05', alpha=0.5)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='#05ffb4', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
tips = pd.read_csv('exemplos/tips.csv')
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
#removo as colunas 1 e 6
party_counts1 = party_counts.loc[:, 2:5]
party_pcts = party_counts.div(party_counts.sum(1), axis=0)
#party_pcts
party_pcts.plot.bar()
```

```python
import seaborn as sns
tips['tip_pct'] = tips['tip']/(tips['total_bill'] - tips['tip'])
tips.head()
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```python
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')
#sns.set(style='whitegrid')
```

- Plotagem de dispersão ou de pontos

```python
marco = pd.read_csv('exemplos/macrodata.csv')
data = marco[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data
```

```python
ax = sns.regplot(x='m1', y='unemp', data=trans_data)
ax.set_title("Changes in log(m1) versus log(unemp)")
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.5})
```

- Grades de facetas e dados de categoria

```python
sns.catplot(x='day', y='tip_pct', hue='time',
            col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x='day', y='tip_pct', row='time',
            col='smoker', kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x='tip_pct', y='day', kind='box',
            data=tips[tips.tip_pct < 0.5])
```
